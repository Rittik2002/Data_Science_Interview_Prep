# Data Visualization Using Seaborn

```python
import seaborn as sns
```

```python
#line plot
y_1=sns.load_dataset("penguins").head(50)
y_1
```

```python
import matplotlib.pyplot as plt
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",size=30)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex")
plt.show()
```

```python
#palette - for colour change
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1,hue="sex",style="sex",palette="Accent")
plt.show()
```

```python
#dash in the female can be removed => dashes=False
#to remove legend from graph set legend=False
#marker can be introdused with markers = list of marker 

sns.lineplot(x="bill_length_mm",y="bill_depth_mm",
             data=y_1,hue="sex",style="sex",palette="Accent",
             dashes=False,markers=["o",">"],legend=False)

plt.grid()#to make grid
plt.title("line plot")
plt.show()
```

# Bar Plot
shows the relationship between a numeric and a categoric variable.

Imagine a barplot showing the average test scores of students in different classes (e.g., Math, Science, English). Each bar would represent a class, and the height of the bar would indicate the average test score for that class, allowing you to see which class had the highest average score. 

```python
df=sns.load_dataset("penguins").head(200)
df
```

```python
df['island'].value_counts()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df,palette='viridis')
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex')
plt.show()
```

```python
#order
order=['Biscoe','Torgersen','Dream']
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',order=order)
plt.show()
```

```python
#order of hue
sns.barplot(x='island',y='bill_length_mm',
            data=df,hue='sex',order=order,
            hue_order=['Female','Male'])
plt.show()
```

```python
#confidence interval-> ci=[0,100]
sns.barplot(x='island',y='bill_length_mm',
            data=df,hue='sex',order=order,
            hue_order=['Female','Male'],ci=96)
plt.show()
```

```python
#confidence interval-> ci=[0,100]
sns.barplot(y='island',x='bill_length_mm',
            data=df,hue='sex',order=order,
            hue_order=['Female','Male'],orient='h',errorbar=('ci',96))
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',
            data=df,hue='sex',order=order,
            hue_order=['Female','Male'],saturation=0.2)
plt.show()
```

```python
#alpha to make light graph
sns.barplot(x='island',y='bill_length_mm',
            data=df,hue='sex',order=order,
            hue_order=['Female','Male'],errorbar=('ci',96),capsize=1,
            err_kws={'color': 'g'},alpha=0.4)
plt.show()
```

# Histogram plot

```python
df=sns.load_dataset('penguins').head(200)
df
```

```python
sns.displot(df["bill_length_mm"])
plt.show()
```

```python
#bins - total number of bins in total range 
#or list may be based - [5,10,15,20]-> 1st bin will be from 5 to 10 then from 10 to 15...
sns.displot(df["bill_length_mm"],bins=15,kde=True)
plt.show()
```

```python
#rug - a visual representation of individual data points along the x-axis
sns.displot(df["bill_length_mm"],kde=True,rug=True,color='g')
plt.show()
```

```python
sns.displot(df["bill_length_mm"],kind='kde')
plt.show()
```

```python
sns.displot(data=df, x="bill_length_mm", kde=True, rug=True, hue="sex")
plt.show()
# grey color represent overlap between two categories
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex',alpha=0.5)
plt.show()
```
