```python
df=sns.load_dataset('anagrams')
```

```python
df
```

```python
df.drop(columns=['attnr'],axis=1,inplace=True)
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df,vmin=0,vmax=10)
```

```python
sns.heatmap(df,vmin=0,vmax=10,annot=True)
```

```python
y={'fontsize':12,'color':'g'}
sns.heatmap(df,vmin=0,vmax=5,annot=True,annot_kws=y,linewidths=4)
```

```python
y={'fontsize':12,'color':'g'}
v=sns.heatmap(df,vmin=0,vmax=5,annot=True,annot_kws=y,
              linewidths=4,cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="python",ylabel='heatmap')
plt.show()
```

# Count plot
plots the count of the number of records by category.
while barplot plots a value(by default mean) for each category

```python
df=sns.load_dataset('tips')
df
```

```python
sns.countplot(x='sex',data=df)
```

```python
sns.barplot(x='sex',y='size',data=df)
```

```python
sns.countplot(x='sex',data=df,hue='smoker',palette='bwr',saturation=0.6)
```

# Violinplot
A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='deep')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='deep',hue='time')
plt.show()
```

```python
sns.violinplot(x='time',y='total_bill',data=df,palette='deep',order=['Dinner','Lunch'])
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='deep',hue='sex',split=True,density_norm='width')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='deep',inner='quart')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=df,palette='deep',inner='box')
plt.show()
```

```python
sns.violinplot(y=df['total_bill'])
plt.show()
```

# pairplot
A pair plot, also known as a scatterplot matrix, is a data visualization tool that helps you understand the relationships between each variables in a dataset.

to plot multiple pairwise bivariate distribution in dataset


```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
plt.show()
```

```python
sns.pairplot(df,hue='sex',y_vars=['total_bill','tip'])
plt.show()
```

```python
sns.pairplot(df,hue='sex',markers=['*','>'],y_vars=['total_bill','tip'],kind='reg')
plt.show()
```

# Stripplot
A strip plot is very simple to understand. It is basically a scatter plot that
differentiates different categories. Strip plots are considered a good
alternative to a box plot or a violin plot.

```python
sns.stripplot(x='day',y='total_bill',data=df,palette='deep')
```

```python
sns.stripplot(x='day',y='total_bill',data=df,palette='deep',hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=df,palette='deep',hue='sex',jitter=10)
```

```python
sns.stripplot(x=df['total_bill'])
```

# Boxplot

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=df)
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=df,hue='time')
plt.show()
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=df,showmeans=True,
            color='w',meanprops={'marker':'*','markeredgecolor':'r'})
plt.show()
```

```python
sns.boxplot(df)
```

# Factorplot
Factor Plot is used to draw a different types of categorical plot . The default plot that is shown is a point plot, but we can plot other seaborn categorical plots by using of kind parameter, like box plots, violin plots, bar plots, or strip plots.

```python
sns.catplot(x='day',y='tip',data=df,kind='point',hue='sex')
```

```python
sns.catplot(x='day',y='tip',data=df,kind='box',hue='sex')
```

```python
sns.catplot(x='day',y='tip',data=df,kind='boxen',hue='sex')
#Boxen, or letter value plots, are an advancement on regular box plots,
#specifically designed to visualize distributions more accurately.
```
