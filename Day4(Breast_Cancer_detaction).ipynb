```python
df['diagnosis']=df['diagnosis'].map({'M':0,'B':1})
```

```python
X=df.drop(columns=['diagnosis'])
y=df['diagnosis']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
y_train.value_counts()
```

```python
# upsampling

from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(random_state=42)
X_resampled,y_resampled=ros.fit_resample(X_train,y_train)
```

```python
y_resampled.value_counts()
```

```python
# SMOTETomek
from imblearn.combine import SMOTETomek
STobj=SMOTETomek(random_state=42)
X_resampled2,y_resampled2=STobj.fit_resample(X_train,y_train)
```

```python
y_resampled2.value_counts()
```

```python
X_resampled2.shape
```

```python
y_test.shape
```

# All classification models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.tree import  DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score, classification_report
```

```python
classifiers={
    'lr':LogisticRegression(max_iter=100),
    'rf':RandomForestClassifier(),
    'gb':GradientBoostingClassifier(),
    'dt':DecisionTreeClassifier(),
    'ab':AdaBoostClassifier(),
    'svm':SVC(),
    'Knn':KNeighborsClassifier(),
    'nb':GaussianNB(),
}
```

```python
import warnings
warnings.filterwarnings('ignore')
for name, clf in classifiers.items():
    clf.fit(X_resampled2,y_resampled2)
    y_pred=clf.predict(X_test)
    print(f"{name}:\n{classification_report(y_test, y_pred)}\n")
```

```python
for name, clf in classifiers.items():
    y_pred=clf.predict(X_resampled2)
    print(f"{name}:\n{classification_report(y_resampled2, y_pred)}\n")
```

best model is adaboost with 96% test accuracy and 100% train accuracy(dataset is not too large)
  

validate whether AdaBoost is truly the best choice

```python
from sklearn.model_selection import cross_val_score
```

```python
ab=AdaBoostClassifier()
scores=cross_val_score(ab,X_resampled2,y_resampled2,scoring='accuracy')
print(scores.mean())
```

```python
ab=AdaBoostClassifier()
scores=cross_val_score(ab,X_test,y_test,scoring='accuracy')
print(scores.mean())
```
